By diagnosing the data frame, we know that:


There are 13 columns of properties with 10841 rows of data.

Column 'Reviews', 'Size', 'Installs' and 'Price' are in the type of 'object'

Values of column 'Size' are strings representing size in 'M' as Megabytes, 'k' as kilobytes and also 'Varies with devices'.


Values of column 'Installs' are strings representing install amount with symbols such as ',' and '+'.

Values of column 'Price' are strings representing price with symbol '$'.

Hence, we will need to do some data cleaning.

```python
# Data cleaning
# 1) Clean the 'Reviews' data and change the type from 'object' to 'float'

def clean_review(review):
    # Replace 'M' with 'million' and convert to float
    if 'M' in review:
        review = float(review.replace('M', '')) * 1_000_000
    return float(review)

# Apply the cleaning function to the 'Reviews' column
apps_df['Reviews'] = apps_df['Reviews'].apply(clean_review)
```

```python
# Data cleaning
# 2) Clean the 'Size' data and change the type from 'object' to 'float'

# Remove rows with '1,000+' in 'Size' as the value is uncertain
apps_df = apps_df[apps_df['Size'] != '1,000+']

def clean_size(size):
    # Convert 'M' to megabytes and 'k' to megabytes
    if 'M' in size:
        size = float(size.replace('M', ''))
    elif 'k' in size:
        size = float(size.replace('k', '')) / 1024  # Convert kilobytes to megabytes
    elif size == 'Varies with device':
        size = 0.0  # Represent 'Varies with device' as 0
    return size

# Apply the cleaning function to the 'Size' column
apps_df['Size'] = apps_df['Size'].apply(clean_size).astype(float)
```

```python
# Data cleaning
# 3) Clean the 'Installs' data and change the type from 'object' to 'float'

def clean_install(install):
    # Remove commas and plus signs, then convert to integer
    install = install.replace(',', '').replace('+', '')
    return int(install)

# Apply the cleaning function to the 'Installs' column
apps_df['Installs'] = apps_df['Installs'].apply(clean_install).astype(float)
```

```python
# Data cleaning
# 4) Clean the 'Price' data and change the type from 'object' to 'float'

def clean_price(price):
    # Remove the dollar sign and convert to float
    return float(price.replace('$', ''))

# Apply the cleaning function to the 'Price' column
apps_df['Price'] = apps_df['Price'].apply(clean_price)
```

```python
# look at the random 10 records in the apps dataframe to verify the cleaned columns
apps_df.sample(10)
```

```python
# check on null values
apps_df.isna().sum()
```

There are 1474 rows having null values under column 'Rating'. Hence, we decided to replace the null values with median of overall 'Rating' values.

```python
# Replace null values in 'Rating' with the median and convert to float
apps_df['Rating'] = apps_df['Rating'].fillna(apps_df['Rating'].median()).astype(float)
```

```python
# Remove rows where 'Type' has a null value
apps_df = apps_df.dropna(subset=['Type'])
```

```python
# check on null values
apps_df.isna().sum()
```

```python
# Group the data by 'App' and calculate the mean of 'Installs'
apps_df = apps_df.groupby(['App'], as_index=False).agg({
    'Reviews': 'first', 
    'Category': 'first', 
    'Rating': 'first', 
    'Size': 'first', 
    'Installs': 'mean', 
    'Type': 'first', 
    'Price': 'first', 
    'Content Rating': 'first', 
    'Genres': 'first', 
    'Last Updated': 'first', 
    'Current Ver': 'first', 
    'Android Ver': 'first'
})

# Sort the dataframe by 'Reviews' in descending order and drop duplicate 'App' entries
apps_df.sort_values(by='Reviews', ascending=False, inplace=True)
apps_df.drop_duplicates(subset=['App'], inplace=True)

apps_df
```

Exploratory Analysis and Visualization

Category

```python
# get the number of apps for each category
sns.set_style('darkgrid')
plt.figure(figsize=(10, 5))
sns.countplot(x='Category', data=apps_df)
plt.title('Number of Apps Per Category')
plt.xticks(rotation=90)
plt.ylabel('Number of Apps')
plt.show()

```

From this plotting we know that most of the apps in the play store are from the categories of 'Family', 'Game' and also 'Tools.

```python
# get the number of installs for each category
categories = apps_df.groupby('Category')
category_installs_sum_df = categories[['Installs']].sum()
category_installs_sum_df = category_installs_sum_df.reset_index()  # to convert groupby object into dataframe

plt.figure(figsize=(10, 5))
sns.barplot(x='Category', y='Installs', data=category_installs_sum_df)
plt.xticks(rotation=90)
plt.ylabel('Installs (e+10)')
plt.title('Number of Installs For Each Category')
plt.show()
```

From this distribution plotting of number of installs for each category, we can see that most of the apps being downloaded and installed are from the categories of 'Game' and 'Communication'.

Rating

```python

# show the distribution of rating
plt.figure(figsize=(10, 5))
sns.countplot(x='Rating', data=apps_df)
plt.title('Rating Distribution')
plt.xticks(rotation=90)
plt.ylabel('Number of Apps')
plt.show()
```

From this distribution plotting, it implies that most of the apps in the Play Store are having rating higher than 4 or in the range of 4 to 4.7.

```python
import matplotlib.pyplot as plt

# Group the data by 'Rating' and sum the other columns
rating_df = apps_df.groupby('Rating').sum().reset_index()

# Create a figure with 4 subplots
fig, axes = plt.subplots(1, 4, figsize=(14, 4))

# Define titles and labels for each subplot
titles = ['Reviews Per Rating', 'Size Per Rating', 'Installs Per Rating', 'Price Per Rating']
columns = ['Reviews', 'Size', 'Installs', 'Price']
colors = ['r', 'g', 'b', 'k']

# Plot each graph
for ax, col, title, color in zip(axes, columns, titles, colors):
    ax.plot(rating_df['Rating'], rating_df[col], color)
    ax.set_xlabel('Rating')
    ax.set_ylabel(col)
    ax.set_title(title)

# Adjust layout and show the plot
plt.tight_layout(pad=2)
plt.show()
```

From the above plottings, we can imply that most of the apps with higher rating range of 4.0 - 4.7 are having high amount of reviews, size, and installs. In terms of price, it doesn't reflect a direct relationship with rating, as we could see a fluctuation in term of pricing even at the range of high rating.
